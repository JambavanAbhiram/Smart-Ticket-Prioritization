```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
# load the dataset

url = r"G:\Machine Learning\IT Log Project\incident_event_log.csv"
file = pd.read_csv(url)
data = pd.DataFrame(file)
print(data.shape)
data.head()
```

PREPROCESSING

DataFrame Manipulation


```python
# dropping unwanted columns
columns_to_drop = ['caller_id', 'opened_at', 'sys_created_at', 'sys_updated_at', 
                   'contact_type', 'cmdb_ci', 'u_priority_confirmation', 'problem_id', 'rfc',
                    'vendor', 'caused_by', 'resolved_at', 'closed_at']
data = data.drop(columns = columns_to_drop, axis = 1)

print(data.shape)
data.head()
```

Data Manipulation

```python
print(data.columns)
```

```python
# replacing spaces with _
columns_to_be_manipulated = ['opened_by', 'sys_created_by', 'sys_updated_by', 'location', 'category', 'subcategory', 'u_symptom', 'impact', 'urgency', 'priority',
                               'assignment_group', 'assigned_to', 'notify', 'closed_code', 'resolved_by']

for cols in columns_to_be_manipulated:
    data[cols] = data[cols].apply(lambda x: x.replace(' ', '_') if isinstance(x, str) else x)

data.head()
```

Text Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

cols_to_be_encoded = ['incident_state', 'active', 'made_sla', 'impact', 'urgency', 'priority', 'knowledge', 'notify']

for cols in cols_to_be_encoded:
    data[cols] = le.fit_transform(data[cols])

data.head()
```

```python
data.replace('?', np.nan, inplace=True)
data.head()
```

```python
import re

cols_to_be_regexed = ['number', 'opened_by', 'sys_created_by', 'sys_updated_by', 'location', 
                      'category', 'subcategory', 'u_symptom', 'assignment_group', 
                      'assigned_to', 'closed_code', 'resolved_by']

def regexing(text):
    if pd.isna(text):
        return text
    if isinstance(text, str):
        match = re.findall(r'\d+', text)
        return int(match[0])

for cols in cols_to_be_regexed:
    data[cols] = data[cols].apply(regexing)

data.head()
```

```python
data.shape
```

```python
# handling null values
for cols in data.columns:
    if pd.api.types.is_numeric_dtype(data[cols]):
        if data[cols].isna().any():
            median = data[cols].median()
            data[cols] = data[cols].fillna(median)

data.head()
```

SPLITTING 

```python
data.columns
```

```python
features = data[['number', 'incident_state', 'active', 'reassignment_count',
       'reopen_count', 'sys_mod_count', 'made_sla', 'opened_by',
       'sys_created_by', 'sys_updated_by', 'location', 'category',
       'subcategory', 'u_symptom', 'impact', 'urgency',
       'assignment_group', 'assigned_to', 'knowledge', 'notify', 'closed_code',
       'resolved_by']]
target = data['priority']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 42)
```

```python
print("Size of training set: ", len(x_train), f"\nSize of test set: ", len(x_test))
```

```python
print(target.value_counts())
```

MODELING AND EVALUATION

```python
from imblearn.over_sampling import SMOTE
```

```python
# importing evaluation metrics
from sklearn.metrics import accuracy_score, classification_report
```

Using Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

model1 = GaussianNB()
smote = SMOTE(random_state = 42)
x_res, y_res = smote.fit_resample(x_train, y_train)
model1.fit(x_res, y_res)
```

```python
y_pred = model1.predict(x_test)
```

```python
print(f"\nAccuracy: ", accuracy_score(y_test, y_pred))
print(f"\nClassification Report: ", f"\n", classification_report(y_test, y_pred))
```

Using XGBOOST

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(objective = 'mullti:softprob',
                    num_class = len(y_train.unique()),
                    eval_metric = 'mlogloss',
                    use_label_encoder = False,
                    random_state = 42)
smote = SMOTE(random_state=42, k_neighbors = 3)
x_res, y_res = smote.fit_resample(x_train, y_train)
xgb.fit(x_res, y_res)
```

```python
prediction = xgb.predict(x_test)
```

```python
print(f"\nAccuracy: ", accuracy_score(y_test, prediction))
print(f"\nClassification Report: ", f"\n", classification_report(y_test, prediction))
```
